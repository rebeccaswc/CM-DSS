## **ðŸš€ Extract specific fields from the log column**

```python
import re

df2 = pd.read_csv("data/clean_data.csv")

def extract_fields(log_entry):
    rule_description = re.search(r'rule\\.description:(.*?)(,|$)', log_entry)
    src_ip = re.search(r'data\\.srcip:(.*?)(,|$)', log_entry)
    mitre_technique = re.search(r'rule\\.mitre\\.technique:\[(.*?)\](,|$)', log_entry)
    rule_id = re.search(r'rule\\.id:(.*?)(,|$)', log_entry)
    fired_times = re.search(r'rule\\.firedtimes:(.*?)(,|$)', log_entry)
    severity_level = re.search(r'rule\\.level:(.*?)(,|$)', log_entry)
    timestamp = re.search(r'timestamp:(.*?)(,|$)', log_entry)

    return {
        'Rule Description': rule_description.group(1).strip() if rule_description else None,
        'Source IP': src_ip.group(1).strip() if src_ip else None,
        'MITRE ATT&CK Technique': mitre_technique.group(1).strip().replace('"', '') if mitre_technique else None,
        'Rule ID': rule_id.group(1).strip() if rule_id else None,
        'Fired Times': fired_times.group(1).strip() if fired_times else None,
        'Severity Level': severity_level.group(1).strip() if severity_level else None,
        'Timestamp': timestamp.group(1).strip() if timestamp else None,
    }

df_extracted = df2['log'].apply(extract_fields)
df_extracted = pd.json_normalize(df_extracted)
df_extracted
```

```python
df_extracted['alertID'] = df_extracted.index + 1

cols = ['alertID'] + [col for col in df_extracted.columns if col != 'alertID']
df_extracted = df_extracted[cols]

print(df_extracted.head().to_string(index=False))
```

```python
df_extracted.to_csv('log_data.csv', index=False)
```
